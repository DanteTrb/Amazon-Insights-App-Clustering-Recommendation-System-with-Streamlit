```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Caricamento dataset
df = pd.read_csv('dataset/amazon_clean.csv')

# Info generali
print("\nüîç INFO GENERALI")
print(df.info())

# Prime righe
print("\nüëÅ PRIME RIGHE")
print(df.head())

# Statistiche descrittive
print("\nüìä STATISTICHE DESCRITTIVE")
print(df.describe(include='all').transpose())

# Valori nulli per colonna
print("\n‚ö†Ô∏è VALORI MANCANTI PER COLONNA")
print(df.isnull().sum())

# Distribuzione delle categorie di prodotto
print("\nüõçÔ∏è DISTRIBUZIONE DELLE CATEGORIE")
print(df['category'].value_counts())

# Distribuzione del rating
plt.figure(figsize=(8, 4))
sns.histplot(df['rating'], bins=20, kde=True)
plt.title('Distribuzione Rating')
plt.xlabel('Rating')
plt.ylabel('Frequenza')
plt.tight_layout()
plt.show()

# Distribuzione prezzi
plt.figure(figsize=(8, 4))
sns.histplot(df['actual_price'], bins=30, kde=True)
plt.title('Distribuzione Prezzo Reale')
plt.xlabel('Prezzo (‚Ç¨)')
plt.ylabel('Frequenza')
plt.tight_layout()
plt.show()
```
